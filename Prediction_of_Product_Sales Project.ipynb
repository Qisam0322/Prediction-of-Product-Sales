## Clean Data

```python
#1) How many rows and columns?
# There are 8523 rows and 12 columns.

#2)What are the datatypes of each variable?
df.dtypes
```

```python
#3) Are there duplicates? If so, drop any duplicates.
df.duplicated()
df.duplicated().sum()
```

```python
#4)Identify missing values.
null_sum=df.isna().sum()
null_percentage=null_sum/len(df)*100
null_percentage
```

```python
null_percentage[null_percentage>0]
```

```python
#5)Address the missing values by using a placeholder value.
df['Item_Weight'].value_counts()
```

```python
#check statistic of the column since there is 17% of null in 'Item_Weight'
df['Item_Weight'].describe()
```

```python
#choose to use -1 as a placeholder because we don't know the weight of
#the items and -1 makes it easy to separate from other
#non-null values.
df['Item_Weight']=df['Item_Weight'].fillna(-1)
df['Item_Weight'].describe()
```

```python
# Check for Null after filling null with 'mean'
df['Item_Weight'].isna().sum()
```

```python
# Inspecting column 'Outlet_Size'.
df['Outlet_Size'].value_counts()

```

```python
# categorical column, decided to use work 'MISSING' as a placeholder to fill Null.
df['Outlet_Size']=df['Outlet_Size'].fillna('MISSING')
df['Outlet_Size'].value_counts()
```

```python
# 6) Confirm that there are no missing values after addressing them.
df.isna().sum()
```

```python
# 7) Find and fix any inconsistent categories of data (for example: fix cat, Cat, and cats so that they are consistent).
```
